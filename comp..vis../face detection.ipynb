```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
nadia=cv2.imread('nadia.jpeg',0)
denis=cv2.imread('denis.jpg',0)
solvay=cv2.imread('solvay1.jpg',0)
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img=img.copy()
    
    face_rects=face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
def adj_detect_face(img):
    
    face_img=img.copy()
    
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),2)
    
    return face_img
```

```python
result=adj_detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
 def detect_eyes(img):
    face_img=img.copy()
    
    eye_rects=eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),2)
    return face_img
```

```python
result=detect_eyes(nadia)
plt.imshow(result,cmap='gray')
#the white of the eye is good so it detect the eye 
```

```python
result=detect_eyes(denis)
plt.imshow(result,cmap='gray')

#because the whites of its eye is really dark then ost of the people
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read(0)
    
    frame=detect_face(frame)
    
    cv2.imshow("VideoFace",frame)
    
    k=cv2.waitKey(1)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()
```
