```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('sudoku.jpeg',0)
```

```python
display_img(img)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx)
```

```python
sobely=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobely)
```

```python
#laplace derivaatives 
laplacian=cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
#blend the sobelx gradient and y gradient 
blended=cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0.5)
```

```python
display_img(blended)
```

```python
ret,th1=cv2.threshold(img,50,255,cv2.THRESH_BINARY)
display_img(th1)
```

```python
kernel=np.ones((4,4),np.uint8)
```

```python
gradient=cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
