```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
def load_img():
    blank_img=np.zeros((600,600))
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300),fontFace=font,fontScale=5,color=(255,0,0),thickness=20)
    return blank_img
```

```python
img=load_img()
display_img(img)
```

```python
#erosion the morphological operator
kernel=np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
#erosion eroding the boundary
result=cv2.erode(img,kernel,iterations=4)
```

```python
display_img(result)
```

```python
# just aded two noises
noise_img=white_noise+img
```
