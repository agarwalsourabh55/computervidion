# corner Harris problem


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
flat_chess=cv2.imread('chess.jpeg')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_class=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_class,cmap='gray')
```

```python
gray=np.float32(gray_flat_class)
```

```python
dst=cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
#dilate is morphological operator
dst=cv2.dilate(dst,None)
```

```python
#taking 1 % of the max value [dst>0.01*dst.max()]
flat_chess[dst>0.01*dst.max()]=[255,0,0] 
```

```python
plt.imshow(flat_chess)
```

# shi tomasi algo 

```python
flat_chess=cv2.imread('chess.jpeg')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
gray_flat_class=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_class,cmap='gray')
```

```python
corners=cv2.goodFeaturesToTrack(gray_flat_class,5,0.01,10)
#5 is giving because to find the 5 corners 
```

```python
#flattened out the array and draw the little circles 
corners=np.int0(corners)
for i in corners:
    x,y=i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners=cv2.goodFeaturesToTrack(gray_flat_class,80,0.01,10)
```

```python
corners=np.int0(corners)
for i in corners:
    x,y=i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```
