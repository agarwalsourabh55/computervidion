```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
full=cv2.imread('puppy.jpeg')
full=cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face=full[30:100,130:200]
plt.imshow(face)
face.shape
```

```python
#all the 6 methods fro comparison in a list
#note how we are using strings ,later on we'll use the eval() function to convert to function
methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for n in methods:
    #create the copy
    full_copy=full.copy()
    
    method=eval(n)
    
    #template matching
    res=cv2.matchTemplate(full_copy,face,method)
    
    min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)
    print(min_val,max_val,min_loc,max_loc)
    
    
    # this is because cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED are using the minimum value for heat matching and
    #other function used max matching
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left=min_loc
    else:
        top_left=max_loc
    
    #this is to make the ractangle on the image 
    height,width,channels=face.shape
    
    bottom_right=(top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),10)
    
    #plot and show the images
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('hatmap of templlate maching')
    
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title("detection of template")
    
    #title with method used
    plt.suptitle(n)
    plt.show()
    
    print('\n')
    print('\n')
    
    
    
```

```python
my_method=eval('cv2.TM_CCOEFF')
res=cv2.matchTemplate(full,face,my_method)
plt.imshow(res)
```
