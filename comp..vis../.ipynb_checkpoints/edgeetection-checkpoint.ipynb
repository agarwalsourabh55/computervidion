```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img=cv2.imread('puppy.jpeg')
```

```python
edges=cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
edges=cv2.Canny(image=img,threshold1=0,threshold2=255)

plt.imshow(edges)
```

```python
med_val=np.median(img)
med_val
```

```python
#LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN CVALUE WHETHER IS GREATER 
lower=int(max(0,0.7*med_val))
#UPPER THRESHOLD TO EITHER 130% OF THE MEDIA  OR THE MAX 255 WHICHEVER IS SAMLLER 
upper=int(min(255,1.3*med_val))
```

```python
edges=cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
upper
```

```python
blurred_img=cv2.blur(img,ksize=(5,5))
```

```python
#probably this will give the best result
# this result show that blurred images show the best edghes 
edges=cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
