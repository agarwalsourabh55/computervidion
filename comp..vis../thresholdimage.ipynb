```python
img=cv2.imread('crossword.jpeg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
```

```python
show_pic(img)
```

```python
ret,th1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
#awe can add any threshold say 127 and 130 etc all fro 0 to 255
```

```python
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)

# just try to hange the attribute to grab the accurate color 
#like cv2,ADAOTIVE change this aattribute to et more accurate answer
```

```python
blended=cv2.addWeighted(src1=th1,alpha=0.6,src2=th2,beta=0.4,gamma=0)
show_pic(blended)
```
