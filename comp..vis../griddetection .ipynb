# grid detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#forthis your image must be clean enough to see the chess board 
flat_chess=cv2.imread('chess.jpeg')
```

```python
found,corners=cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
found
```

```python
#corners
```

```python
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
```

```python
plt.imshow(flat_chess)
```

```python
found,corners=cv2.findCirclesGrid(flat_chess,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
```

```python
plt.imshow(flat_chess)
```
