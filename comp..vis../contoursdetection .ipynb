# Contour Detection 


```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img=plt.imread('smiley.jpeg',0)
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img
```

```python
plt.imshow(img,cmap='gray')
```

```python
contours,hierarchy=cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
#cv2.RETR_CCOMP it extrxt both internal and external
#cv2.CHANGE_APPROX_SIMPLE it return image and hierarchy 

#this must return three but now this is returning two values so have problem with this  
```

```python
#print(image)
```

```python
#len(contours)
#type(hierarchy) # a numpy array

external_contours=np.zeros(img.shape)
external_contours.shape

for i in range(len(contours)):
    
    #external contours if ==-1
    if hierarchy[0][1][3]== -1:
        
        cv2.drawContours(external_contours,contours,i,255,-1)
      
```

```python
plt.imshow(external_contours,cmap='gray')
```

```python
internal_contours=np.zeros(image.shape)

for i in range(len(contours)):
    
    #internal contours if !=-1
    if hierarchy[0][i][3]!= -1:
        
        cv2.drawContours(external_contours,contours,i,255,-1)
      
```
