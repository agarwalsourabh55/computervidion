```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    img=cv2.imread('bricks.jpeg').astype(np.float32)/255
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img 
```

```python
#load_img()
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img)
```

```python
i=load_img()
display_img(i)
```

```python
#gamma correction
gamma=1/4 #gamma less then 1 is brighter the image 
#gamma above 1 image appear will be darker 
```

```python
result=np.power(i,gamma)
display_img(result)
```
