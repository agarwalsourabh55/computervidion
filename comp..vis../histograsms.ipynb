```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dark_horse=cv2.imread('horse.jpeg') #original BGR channeling for =opencv 
show_horse=cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)#converted to RGB
rainbow=cv2.imread('rainbow.jpeg')
rainbow=cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
bricks=cv2.imread('bricks.jpeg')
bricks=cv2.cvtColor(bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rainbow)
```

```python
#OPENCV BGR
hist_values=cv2.calcHist([bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
hist_values.shape
```

```python
plt.imshow(hist_values)
```

```python
img=bricks
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('histogram for bricks')
#changing th x limit would affect the answer 
#we can also add ylimit if we waanted to add
```
